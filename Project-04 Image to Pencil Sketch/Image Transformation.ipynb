# Image To Pencil Sketch 

### Introduction to Transforming Images: From RGB to Pencil Sketch Effect
In image processing, there are various techniques available to transform and manipulate images in order to achieve desired effects. One such technique involves converting an image from RGB format to grayscale, followed by inverting it to create a negative image. Additionally, combining the grayscale image with an inverted blurry image can generate a pencil sketch effect. These operations can be performed programmatically using the divide function from the cv2 library in Python.

The process begins by reading the image in RGB format, which consists of red, green, and blue color channels. Converting the image to grayscale effectively removes color information and creates a black and white representation. This step is commonly used to simplify images and focus on their structural details.

Once the grayscale image is obtained, it can be inverted to create a negative image. Inversion involves reversing the brightness values of the pixels, resulting in an image with inverted tones. This technique can be useful in enhancing details and highlighting certain features within the image.

To achieve a pencil sketch effect, the grayscale image is combined with an inverted blurry image. The blurry image helps to smooth out the details and create a blending effect when mixed with the grayscale image. The process of blending the two images can be performed by dividing the grayscale image by the inverted blurry image using the divide function from the cv2 library.

By employing these steps, we can transform a regular image into a classic black and white photo and further enhance it by creating a pencil sketch effect. The ability to manipulate images programmatically opens up creative possibilities and allows for the generation of visually appealing and artistic results.

```python
from sklearn.decomposition import PCA, IncrementalPCA
pca=PCA()
pca.fit(ib)
var=np.cumsum(pca.explained_variance_ratio_)*100
k=np.argmax(var>95)
print("elements are "+str(k)) 
```

```python
plt.figure(figsize=(10,8))
plt.ylabel('variance')
plt.xlabel('PCA')
plt.axvline(x=k,color='b',linestyle='--')
plt.axhline(y=95,color='g',linestyle='--')
ax=plt.plot(var)
```

```python
ipca=IncrementalPCA(n_components=2)
i=ipca.inverse_transform(ipca.fit_transform(ib))
plt.figure(figsize=(6,4))
plt.imshow(i)
```

```python
from mpl_toolkits.mplot3d import Axes3D
import cv2
o_img=cv2.imread('./photos/m-044-01.jpg')
img=cv2.cvtColor(o_img,cv2.COLOR_BGR2RGB)
v=img.reshape(-1,3)
plt.imshow(o_img)
```

```python
import numpy as np
v=np.float32(v)
criteria=(cv2.TERM_CRITERIA_EPS+cv2.TERM_CRITERIA_MAX_ITER,10,1.0)
k=3
attempts=10
ret,label,center=cv2.kmeans(v,k,None,criteria,attempts,cv2.KMEANS_PP_CENTERS)
center=np.uint8(center)
res=center[label.flatten()]
r=res.reshape(img.shape)
f1=6
f2=4
plt.figure(figsize=(f1,f2))
plt.subplot(1,2,1),plt.imshow(img)
plt.subplot(1,2,2),plt.imshow(r)
```

```python
gray_image=cv2.cvtColor(image_raw,cv2.COLOR_BGR2GRAY) ##convert to grey image
plt.imshow(gray_image)
```

```python
inverted_image=255-gray_image ## converting to invert image
plt.imshow(inverted_image)
```

```python
## gaussian Blur image
gblur_image1=cv2.GaussianBlur(inverted_image,(5,5),0) ##kernel size
plt.imshow(gblur_image1)
```

```python
## inverted gaussian blur image
invgblur_image=255-gblur_image1
plt.imshow(invgblur_image)
```

```python
blur_image=255-inverted_image
p_image=255-blur_image
pencil_sketch1=cv2.divide(gray_image,p_image,scale=256.0)
plt.imshow(pencil_sketch1)
```

```python
sketch_img=cv2.divide(gray_image,invgblur_image,scale=256.0)
plt.axis(False)
plt.imshow(sketch_img)
```

```python
#original image vs Pencil sketch image comparison
plt.subplot(1,2,1)
plt.title('This is Original image',size=10)
plt.imshow(image_raw)
plt.axis('off')
plt.subplot(1,2,2)
plt.title('This is pencil sketch',size=10)
plt.imshow(sketch_img)
plt.axis('off')
plt.show()
```

```python
plt.title('This is rgb sketch image',size=10)
sketch=cv2.cvtColor(sketch_img,cv2.COLOR_BGR2RGB)
plt.imshow(sketch)
plt.axis('off')
plt.show()
```

The combination of the grayscale image with an inverted blurry image creates a pencil sketch effect by blending the two together.
